```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
# Set the URL you want to webscrape from
url = 'http://www.nuforc.org/webreports/ndxevent.html'
```

```python
# Connect to the URL
response = requests.get(url)
```

```python
# Test response
response
```

```python
# Parse HTML and save to BeautifulSoup objectÂ¶
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Find links
soup.findAll('a')
```

```python
one_a_tag = soup.findAll('a')[1]
link = one_a_tag['href']
```

```python
link
```

```python
# To download the whole data set, let's do a for loop through all a tags
for i in range(1,115): #'a' tags are for links
    one_a_tag = soup.findAll('a')[i]
    link = one_a_tag['href']
    download_url = 'http://www.nuforc.org/webreports/'+ link
    urllib.request.urlretrieve(download_url,'./'+link[link.find('/ndxe_')+1:]) 
    time.sleep(2) #pause the code for a sec
```

```python
# Combine text files into one
filenames = ['turnstile_190504.txt','turnstile_190511.txt']
with open('/Users/ashwingovindaraj/Desktop/Python Udemy/Combinedfile11', 'w') as outfile:
    for fname in filenames:
        with open(fname) as infile:
            outfile.write(infile.read())
```
